### Introduction of the problem:
The objective of this project is to analyze customer satisfaction scores, which obtained from a feedback from over 120,000 airline passengers. The primary goal is to gain insights into the key factors influencing customer satisfaction in the airline industry. These are the questions to address:
 - What are the key factors that contribute to customer satisfaction in the airline industry?
 - Which specific factors (cleanliness, comfort, service, etc.) are the most significant in determining overall satisfaction?
 - How do different aspects of the flight, such as duration, class, and type of travel, influence passenger satisfaction?
 - Can we identify patterns indicating specific preferences within different types of travel. For example, the most consideration of satisfaction for business could be services and the most consideration of satisfaction for personal could be comfort.

The project aims to provide actionable insights to stakeholders in the airline industry that are able to enhance customer satisfaction, tailor services to different traveler types, and ultimately improve the overall travel experience for passengers.


### Data info: 
This data set contains 24 columns and 129,881 records. The major columns include type of airline customer (first-time or returning), purpose of the flight (business or personal), type of passenger seat, flight distance, fight delay information, the sentiment of passengers in term of time, online booking, check-in service, seat comfort, cleanliness, in-flight services, food and drink, facilitation, baggage handling, and overall satisfaction.

### Import libraries and Data loading

```python
# Import all libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import linear_model
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression 
from sklearn.metrics import mean_squared_error, classification_report
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import VarianceThreshold, SelectFromModel, RFE 
import statsmodels.api as sm
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

%matplotlib inline

#Data loading
file_path = 'C:/Users/admin/Desktop/Project/Dataset/airline_passenger_satisfaction.csv'
df = pd.read_csv(file_path)
df.head()
```

## Data Preparation


### Data Preparation
1.Record Selection: due to dataset has over 100,000 that could affect the long-time loading process. Therefore, It is necessary to recduce the number of records while random selction as the same proportion of overall result, which choose gender, type of class and satisfaction rely on the overall proportion of dataset.\
2.Check missing value and fill the average value\
3.Check the duplicated records and remove them\
4.Remove outliner for Departure Delay column and Arrival Delay column\
5.Drop a unique column and transform categorical columns into numerical value

```python
#Records selction 
total_rows = 2000

female_percentage = 0.51
male_percentage = 0.49

business_percentage = 0.5
economy_percentage = 0.5

satisfied_percentage = 0.43
dissatisfied_percentage = 0.57

female_rows = int(total_rows * female_percentage)
male_rows = total_rows - female_rows

business_rows = int(total_rows * business_percentage)
economy_rows = total_rows - business_rows

satisfied_rows = int(total_rows * satisfied_percentage)
dissatisfied_rows = total_rows - satisfied_rows

# Select random rows for each category
female_sample = df[df['Gender'] == 'Female'].sample(female_rows, random_state=42, replace=True)
male_sample = df[df['Gender'] == 'Male'].sample(male_rows, random_state=42, replace=True)

business_sample = df[df['Class'] == 'Business'].sample(business_rows, random_state=42, replace=True)
economy_sample = df[df['Class'] == 'Economy'].sample(economy_rows, random_state=42, replace=True)

satisfied_sample = df[df['Satisfaction'] == 'Satisfied'].sample(satisfied_rows, random_state=42, replace=True)
neutral_dissatisfied_sample = df[df['Satisfaction'] == 'Neutral or Dissatisfied'].sample(dissatisfied_rows, random_state=42, replace=True)

# Combining the samples to get the final subset
final_subset = pd.concat([female_sample, male_sample, business_sample, economy_sample, satisfied_sample, neutral_dissatisfied_sample])

# Shuffle records
df_sample = final_subset.sample(frac=1, random_state=42).reset_index(drop=True)
df_sample.info()
```

```python
#Check NA
print("Before: Total missing value of each column:\n{}".format(df_sample.isna().sum()))
print()
#fill NA with the average value
df_sample.fillna(df_sample.mean(), inplace=True)
print("After: Total missing value of each column:\n{}".format(df_sample.isna().sum()))
```

```python
#check duplicate data
print("Before removing the duplicated data")
print("The duplicated records contain:", df_sample.duplicated().sum(),"records")
df_sample.drop_duplicates(keep='first', inplace = True)
print("After removing the duplicated data")
print("The duplicated records contain:", df_sample.duplicated().sum(),"records")
```

```python
#Observing the Overall information
df_sample.info()
df_sample.describe()
#It's obvious that data analytics display the zero value of first quartile and second 
#for Departure Delay column and Arrival Delay column 
```

```python
#Observing the distribution of Flight Distance, Departure Delay and Arrival Delay
df_selCol = pd.DataFrame(df_sample, columns=['Departure Delay','Arrival Delay'])
df_selCol.boxplot()
print()

#Remove outliner
df_sample.drop(df_sample.loc[df_sample['Departure Delay']>500].index, inplace=True)
df_sample.drop(df_sample.loc[df_sample['Arrival Delay']>500].index, inplace=True)
```

```python
#Drop ID columns because it's unique
df_sample.drop('ID', axis=1, inplace=True)

#Transform Gender, Customer Type, Type of Travel, Class and Satisfaction columns into numerical value
# Create a mapping dictionary
gender_mapping = {'Male': 1, 'Female': 0}
customerType_mapping = {'First-time': 1, 'Returning': 0}
travelType_mapping = {'Business': 1, 'Personal': 0}
class_mapping = {'Business': 2, 'Economy Plus': 1, 'Economy':0}
satisfaction_mapping = {'Satisfied': 1, 'Neutral or Dissatisfied': 0}

# Apply maping
df_sample['Gender'] = df['Gender'].map(gender_mapping)
df_sample['Customer Type'] = df['Customer Type'].map(customerType_mapping)
df_sample['Type of Travel'] = df['Type of Travel'].map(travelType_mapping)
df_sample['Class'] = df['Class'].map(class_mapping)
df_sample['Satisfaction'] = df['Satisfaction'].map(satisfaction_mapping)

df_sample.head(20)
```

### Features selection
We will try different 4 methods of feature selction. 
After observing the name of columns and test with logistic model in different result, the method4 is selected because it includes high correlation of features and interesting columns and provide high accuracy as the method2 and the method3  

```python
#define respone (x) and response (y)
X = df_sample.drop(['Satisfaction'], axis=1)
y = df_sample['Satisfaction']

#Method 1: using variance treshold
vt = VarianceThreshold(threshold=0.25)
vt.fit(X)
pd.DataFrame({'variance': vt.variances_,\
             'select_feature':vt.get_support()}, index = X.columns)
X_1 = X.iloc[:, vt.get_support()]
X_1.shape
```

```python
#Method 2: Use SelectFromModel and LogisticRegression
method2 = SelectFromModel(estimator=LogisticRegression(max_iter=5000))
method2.fit(X,y)

pd.DataFrame({'coef':method2.estimator_.coef_[0],\
             'select_feature':method2.get_support()}, index=X.columns)
X_2 = method2.transform(X)
X_2.shape
```

```python
#Method 3: Use Generalized Linear Model (Binomial) from statsmodels
logit_model = sm.GLM(y,X,family=sm.families.Binomial())
result = logit_model.fit(fit_intercept=True)
print(result.summary2())

#Select the features that coefficient value is greater than 0.02
X_3 = X.loc[:,abs(result.params)>=0.1]
X_3.shape
```

```python
#Method 4: Use Recursive Feature Elimination (RFE) and Logistic Regression
rfe = RFE(estimator=LogisticRegression(max_iter=5000),\
         n_features_to_select=10, step=1)
rfe_result = rfe.fit(X,y)
print(rfe_result.support_)
print(rfe_result.ranking_)

X_4 = X.loc[:,rfe_result.support_]
X_4.head()
```

## Apply model

### 1.SVM

```python
#Select method 4 of feature selection
X_sel = X_4

# split data into train+validation set and test set
X_trainval, X_test, y_trainval, y_test = train_test_split(X_sel, y, random_state=42)

# split train+validation set into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X_trainval, y_trainval, random_state=42)
print("Size of training set: {} size of validation set: {} size of test set:"
      " {}\n".format(X_train.shape[0], X_valid.shape[0], X_test.shape[0]))
      

```

```python
#Create GridSerach with pipeline to optimize the hyperparameters for SVM
pipe = Pipeline([('scaler', StandardScaler()),('svc',SVC())])

grid_parameters = {'svc__kernel': ['linear', 'rbf'],
                   'svc__gamma': [0.01, 0.1],
                   'svc__C': [0.1, 1],
                   'scaler__with_mean': [True, False]}

#Using Pipe
clf = GridSearchCV(pipe, grid_parameters, cv=5, scoring='accuracy')
clf.fit(X_train, y_train)

means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']

for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.5f (%0.05f) for %r" % (mean, std, params))

print("Best Parameter:",clf.best_params_)
print("Best Train score:",clf.best_score_)

gs_best_svm = clf.best_estimator_
score_test_svm = gs_best_svm.score(X_test, y_test)
print("Best Test score:", score_test_svm)

```

### 2.Decision Tree

```python
X_train, X_test, y_train, y_test = train_test_split(X_sel, y, stratify=y, random_state=42)

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train,y_train)

train_acc = tree.score(X_train, y_train)
test_acc = tree.score(X_test, y_test)

print("Decision Tree without max-dept control")
print("Original Train accuracy: ",train_acc)
print("Original Test accuracy: ",test_acc)
print()

#Create GridSerach with pipeline to optimize the hyperparameters for decision tree
#We set the max_dept (manipulate the level of tree for mitigating the overfitting)
pipe = Pipeline([('scaler', StandardScaler()),('tree',DecisionTreeClassifier())])

grid_parameters = {'tree__max_depth': [2, 3, 4, 5],
                   'scaler__with_mean': [True, False]}

#Using Pipe
clf = GridSearchCV(pipe, grid_parameters, cv=5, scoring='accuracy')
clf.fit(X_train, y_train)

means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']

for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.5f (%0.05f) for %r" % (mean, std, params))

print("Best Parameter:",clf.best_params_)
print("Best Train score:",clf.best_score_)

gs_best_tree = clf.best_estimator_
score_test_tree = gs_best_tree.score(X_test, y_test)
print("Best Test score:", score_test_tree)
```

```python
#Decision tree explanation
tree2 = DecisionTreeClassifier(random_state=42, max_depth=3)
tree2.fit(X_train, y_train)

featuresName = feature_names=["Gender", "Customer Type","Type of Travel","Class","On-board Service","Seat Comfort",\
                              "Cleanliness","In-flight Service","In-flight Wifi Service","In-flight Entertainment" ]

export_graphviz(tree2, out_file="tree2.dot", 
                class_names=["malignant", "benign"],
                feature_names= featuresName, 
                impurity=False,                      
                filled=True, rounded=True)

!pip install graphviz
import graphviz
tree_plot2 = graphviz.Source.from_file("tree2.dot")
tree_plot2
```

### 3.Random Forest

```python
# Decision Tree without max-depth control
tree = RandomForestClassifier(random_state=42)
tree.fit(X_train, y_train)

train_acc = tree.score(X_train, y_train)
test_acc = tree.score(X_test, y_test)

print("Random Forest without max-depth control")
print("Original Train accuracy: ", train_acc)
print("Original Test accuracy: ", test_acc)
print()

# Create GridSearch with pipeline to optimize the hyperparameters for Random Forest
pipe_rf = Pipeline([('scaler', StandardScaler()), ('rf', RandomForestClassifier())])

grid_parameters_rf = {'rf__n_estimators': [50, 100, 150],
                      'rf__max_depth': [3, 5, 10],
                      'scaler__with_mean': [True, False]}

# Using Pipe for Random Forest
clf_rf = GridSearchCV(pipe_rf, grid_parameters_rf, cv=5, scoring='accuracy')
clf_rf.fit(X_train, y_train)

means_rf = clf_rf.cv_results_['mean_test_score']
stds_rf = clf_rf.cv_results_['std_test_score']

for mean, std, params in zip(means_rf, stds_rf, clf_rf.cv_results_['params']):
    print("%0.5f (%0.05f) for %r" % (mean, std, params))

print("Best Parameters:", clf_rf.best_params_)
print("Best Train score:", clf_rf.best_score_)

gs_best_rf = clf_rf.best_estimator_
score_test_rf = gs_best_rf.score(X_test, y_test)
print("Best Test score:", score_test_rf)
```

### 4.K-Nearest Neighbors (KNN)

```python
# Decision Tree without max-depth control
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

train_acc = knn.score(X_train, y_train)
test_acc = knn.score(X_test, y_test)

print("K-Nearest Neighbors without hyperparameter tuning")
print("Original Train accuracy: ", train_acc)
print("Original Test accuracy: ", test_acc)
print()

# Create GridSearch with pipeline to optimize the hyperparameters for K-Nearest Neighbors
pipe_knn = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier())])

grid_parameters_knn = {'knn__n_neighbors': [3, 4, 5],
                       'knn__weights': ['uniform', 'distance'],
                       'scaler__with_mean': [True, False]}

# Using Pipe for K-Nearest Neighbors
clf_knn = GridSearchCV(pipe_knn, grid_parameters_knn, cv=5, scoring='accuracy')
clf_knn.fit(X_train, y_train)

means_knn = clf_knn.cv_results_['mean_test_score']
stds_knn = clf_knn.cv_results_['std_test_score']

for mean, std, params in zip(means_knn, stds_knn, clf_knn.cv_results_['params']):
    print("%0.5f (%0.05f) for %r" % (mean, std, params))

print("Best Parameters:", clf_knn.best_params_)
print("Best Train score:", clf_knn.best_score_)

gs_best_knn = clf_knn.best_estimator_
score_test_knn = gs_best_knn.score(X_test, y_test)
print("Best Test score:", score_test_knn)
```

### 5.Logististic Regression

```python
# Logistic Regression without hyperparameter tuning
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)

train_acc = log_reg.score(X_train, y_train)
test_acc = log_reg.score(X_test, y_test)

print("Logistic Regression without hyperparameter tuning")
print("Original Train accuracy: ", train_acc)
print("Original Test accuracy: ", test_acc)
print()

# Create GridSearch with pipeline to optimize the hyperparameters for Logistic Regression
pipe_lr = Pipeline([('scaler', StandardScaler()), ('lr', LogisticRegression())])

grid_parameters_lr = {'lr__C': [0.01, 0.1, 1, 10],
                      'scaler__with_mean': [True, False]}

# Using Pipe for Logistic Regression
clf_lr = GridSearchCV(pipe_lr, grid_parameters_lr, cv=5, scoring='accuracy')
clf_lr.fit(X_train, y_train)

means_lr = clf_lr.cv_results_['mean_test_score']
stds_lr = clf_lr.cv_results_['std_test_score']

for mean, std, params in zip(means_lr, stds_lr, clf_lr.cv_results_['params']):
    print("%0.5f (%0.05f) for %r" % (mean, std, params))

print("Best Parameters:", clf_lr.best_params_)
print("Best Train score:", clf_lr.best_score_)

gs_best_lr = clf_lr.best_estimator_
score_test_lr = gs_best_lr.score(X_test, y_test)
print("Best Test score:", score_test_lr)
```
