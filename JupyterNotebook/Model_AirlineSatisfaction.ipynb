# Term Project Report:  The classification of Airline Passenger Satisfaction
### Fundamental of Machine Learning 




### Introduction of the problem:
The objective of this project is to analyze customer satisfaction scores, which obtained from a feedback from over 120,000 airline passengers. The primary goal is to gain insights into the key factors influencing customer satisfaction in the airline industry. These are the questions to address:
 - What are the key factors that contribute to customer satisfaction in the airline industry?
 - Which specific factors (cleanliness, comfort, service, etc.) are the most significant in determining overall satisfaction?
 - How do different aspects of the flight, such as duration, class, and type of travel, influence passenger satisfaction?
 - Can we identify patterns indicating specific preferences within different types of travel. For example, the most consideration of satisfaction for business could be services and the most consideration of satisfaction for personal could be comfort.

The project aims to provide actionable insights to stakeholders in the airline industry that are able to enhance customer satisfaction, tailor services to different traveler types, and ultimately improve the overall travel experience for passengers.


### Data info: 
This data set contains 24 columns and 129,881 records. The major columns include type of airline customer (first-time or returning), purpose of the flight (business or personal), type of passenger seat, flight distance, fight delay information, the sentiment of passengers in term of time, online booking, check-in service, seat comfort, cleanliness, in-flight services, food and drink, facilitation, baggage handling, and overall satisfaction. 
        


### Import libraries and Data loading

```python
# Import all libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import linear_model
from sklearn.linear_model import LinearRegression, Ridge, Lasso, LogisticRegression 
from sklearn.metrics import mean_squared_error, accuracy_score, log_loss, classification_report, roc_curve, auc, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, RobustScaler
from sklearn.feature_selection import VarianceThreshold, SelectFromModel, RFE 
import statsmodels.api as sm
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
import time
import graphviz

%matplotlib inline

# Project runtime
project_start_time = time.time()

#Data loading
df = pd.read_csv('airline_passenger_satisfaction.csv')
df.head()
```

## Data Preparation and Wrangling


1. Record Selection: due to dataset has over 100,000 that could affect the long-time loading process. Therefore, It is necessary to recduce the number of records while random selction as the same proportion of overall result, which choose gender, type of class and satisfaction rely on the overall proportion of dataset.\
1. Create dummy variables.
1. Handle missing value and duplicates.
1. Apply scaling with RobustScaler.
1. Remove outliers

```python
#Records selection 
total_rows = 2000

female_percentage = 0.51
male_percentage = 0.49

business_percentage = 0.5
economy_percentage = 0.5

satisfied_percentage = 0.43
dissatisfied_percentage = 0.57

female_rows = int(total_rows * female_percentage)
male_rows = total_rows - female_rows

business_rows = int(total_rows * business_percentage)
economy_rows = total_rows - business_rows

satisfied_rows = int(total_rows * satisfied_percentage)
dissatisfied_rows = total_rows - satisfied_rows

# Select random rows for each category
female_sample = df[df['Gender'] == 'Female'].sample(female_rows, random_state=42, replace=True)
male_sample = df[df['Gender'] == 'Male'].sample(male_rows, random_state=42, replace=True)

business_sample = df[df['Class'] == 'Business'].sample(business_rows, random_state=42, replace=True)
economy_sample = df[df['Class'] == 'Economy'].sample(economy_rows, random_state=42, replace=True)

satisfied_sample = df[df['Satisfaction'] == 'Satisfied'].sample(satisfied_rows, random_state=42, replace=True)
neutral_dissatisfied_sample = df[df['Satisfaction'] == 'Neutral or Dissatisfied'].sample(dissatisfied_rows, random_state=42, replace=True)

# Combining the samples to get the final subset
final_subset = pd.concat([female_sample, male_sample, business_sample, economy_sample, satisfied_sample, neutral_dissatisfied_sample])

# Shuffle records
df_sample = final_subset.sample(frac=1, random_state=42).reset_index(drop=True)
df_sample.info()
```

```python
# Create dummy variables for categorical features
df_sample = pd.get_dummies(df_sample, drop_first=True)
```

### Features selection
We will try different 5 methods of feature selction. 
After observing the name of columns and test with logistic model in different result, the method5 is selected.

```python
#define X and y
X = df_sample.drop(['Satisfaction_Satisfied'], axis=1)
y = df_sample['Satisfaction_Satisfied']

# Satisfaction_Neutral or Dissatisfied	Satisfaction_Satisfied

#Method 1: using variance threshold
vt = VarianceThreshold(threshold=0.25)
vt.fit(X)
pd.DataFrame({'variance': vt.variances_,\
             'select_feature':vt.get_support()}, index = X.columns)
X_1 = X.iloc[:, vt.get_support()]
X_1.shape
```

```python
#Method 2: Use SelectFromModel and LogisticRegression
method2 = SelectFromModel(estimator=LogisticRegression(max_iter=5000))
method2.fit(X,y)

pd.DataFrame({'coef':method2.estimator_.coef_[0],\
             'select_feature':method2.get_support()}, index=X.columns)
X_2 = method2.transform(X)
X_2.shape
```

```python
#Method 3: Use Generalized Linear Model (Binomial) from statsmodels
logit_model = sm.GLM(y,X,family=sm.families.Binomial())
result = logit_model.fit(fit_intercept=True)
print(result.summary2())

#Select the features that coefficient value is greater than 0.02
X_3 = X.loc[:,abs(result.params)>=0.1]
X_3.shape
```

```python
#Method 4: Use Recursive Feature Elimination (RFE) and Logistic Regression
rfe = RFE(estimator=LogisticRegression(max_iter=5000),\
         n_features_to_select=10, step=1)
rfe_result = rfe.fit(X,y)
print(rfe_result.support_)
print(rfe_result.ranking_)

X_4 = X.loc[:,rfe_result.support_]
X_4.shape
```

```python
#Method 5: Use SelectFromModel and Random Forest classifier
method5 = SelectFromModel(estimator=RandomForestClassifier(random_state=42, n_estimators=100))
method5.fit(X, y)

X_5 = method5.transform(X)
print(method5.estimator_.feature_importances_)
print(method5.get_support(indices=True))
X_5.shape
```

## Apply model

```python
# Create train, validation and test set
X_sel = X_5
# split data into train+validation set and test set
X_trainval, X_test, y_trainval, y_test = train_test_split(X_sel, y, test_size=0.3, random_state=42)
# split train+validation set into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X_trainval, y_trainval, test_size=0.3, random_state=42)
```

```python
# Apply classifiers
#Names of the various classification approaches for easy presentation of the results
names = ["Logististic Regression", "Nearest Neighbors", "Linear SVM", "RBF SVM",          
         "Decision Tree", "Naive Bayes", "Random Forest"]

scores = [] #list variable to hold classification scores
classifier_scores = {} #dictionary variable to store the classifier name and score

classifiers = [
    LogisticRegression(),
    KNeighborsClassifier(3),
    SVC(kernel='linear'),
    SVC(kernel='rbf'),
    DecisionTreeClassifier(max_depth=3),
    GaussianNB(),
    RandomForestClassifier()
]

for classifier, name in zip(classifiers, names):
    pipeline = Pipeline(steps=[
                ('classifier', classifier)
        ])
    pipeline.fit(X_train, y_train)
#     print(classifier)
    score = pipeline.score(X_test, y_test)
    scores.append(score)
    classifier_scores[name] = score
#     print(score)
    
scores_df = pd.DataFrame(zip(names, scores), \
                         columns=['Classifier', 'Accuracy'])
print(scores_df)

# Print the classifier with the best accuracy
best_classifier = max(classifier_scores, key=classifier_scores.get)
print()
print(f"Best Classifier: {best_classifier}")
print(f"Accuracy: {classifier_scores[best_classifier]}")
```

### Hyperparameter tuning and Grid Search with CV

```python
# start time
start_time_tuning = time.time()

best_score = 0
gamma_list = [0.5,0.95,1]
C_list = [5,6,7]

for gamma in gamma_list:
    for C in C_list:
        svm = SVC(kernel='rbf', gamma=gamma, C=C)
        scores = cross_val_score(svm, X_trainval, y_trainval, cv=5)
        score = np.mean(scores)
        if score > best_score:
            best_score = score
            best_hyperparams = {'C': C, 'gamma': gamma}

print("Best score:", best_score)
print("Best hyperparameters:", best_hyperparams)

tuned_svm = SVC(kernel='rbf', **best_hyperparams)
tuned_svm.fit(X_trainval, y_trainval)

score_test = tuned_svm.score(X_test, y_test)
print("Test score:", score_test)

# end time
end_time_tuning = time.time()

# Calculate the runtime
runtime_tuning = end_time_tuning - start_time_tuning
print("Runtime for hyperparameter tuning:", runtime_tuning, "seconds")
```
